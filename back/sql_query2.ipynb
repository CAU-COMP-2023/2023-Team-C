# SQL query 생성 코드 2

엑셀 파일을 불러온 후, 자동으로 SQL 쿼리로 변환해주는 코드입니다. \
(메뉴 한정)

```python
import pandas as pd

# 데이터 불러오기
df = pd.read_excel('../refined_menu.xlsx')
```

```python
df.head(3)
```

```python
df_new = pd.DataFrame(columns=df.columns)
```

```python
for _, restaurant, menus, prices in df.itertuples():
    menu_list: list[str] = menus.split(',')
    price_list: list[str] = prices.split(',')

    if len(menu_list) != len(price_list):
        print(f'{restaurant}의 메뉴 목록과 가격 목록이 개수가 같지 않습니다! 확인해주세요.')

    for menu, price in zip(menu_list, price_list):
        df_new.loc[len(df_new.index)] = [restaurant, menu.strip(), int(price)]
```

```python
df_new
```

```python
# 쿼리 제작
print('INSERT INTO [dbo].[Menus] VALUES')
for _, restaurant, menu, price in df_new.itertuples():
    print(f"(N'{restaurant}', N'{menu}', {price}),")
print(';')    
```
